```python
import qutip
import numpy as np
b = qutip.Bloch()

b.make_sphere()

z = (qutip.basis(2, 0) + (0+0j)*qutip.basis(2, 1)).unit()
b.add_states([z])
b.render()
b.show()


b = qutip.Bloch()

b.make_sphere()

z = (qutip.basis(2, 0)  (0+0j)*qutip.basis(2, 1)).unit()
b.add_states([z])
b.render()
b.show()
```
